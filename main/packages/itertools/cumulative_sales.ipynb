#### Cumulative sales (accumulate example)


```python
""" Understand how sales are growing over time and to identify any trends.
"""

from itertools import accumulate

# Daily sales data for a week
daily_sales = [120, 180, 240, 300, 560, 420, 480]

# User accumulate
cumulative_sales = list(accumulate(daily_sales))
    # print(cumulative_sales)

# We can plot this to visualize the trend
import matplotlib.pyplot as plt
plt.plot(cumulative_sales, marker='o')
plt.title('Cumulative Sales Over the Week')
plt.xlabel('Day')
plt.ylabel('Cumulative Sales')
plt.grid(True)
```
